```python
#Imports
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
#Load the model from Drive
model_path = "/content/drive/MyDrive/multilabel_cnn_improved.keras"

model = load_model(model_path)
print("Model loaded successfully from Google Drive")
```

```python
#Generate Tuning log
tuning_log = {
    "Model Type": model.__class__.__name__,
    "Total Layers": len(model.layers),
    "Output Activation": model.layers[-1].activation.__name__,
    "Loss Function": model.loss,
    "Optimizer": model.optimizer.__class__.__name__,
    "Learning Rate": round(float(model.optimizer.learning_rate.numpy()), 6)
}

print("MODEL-DERIVED TUNING LOG\n")
for k, v in tuning_log.items():
    print(f"{k}: {v}")
```
